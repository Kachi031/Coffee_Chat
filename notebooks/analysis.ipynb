```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import coffee data
coffee = pd.read_csv("../data/crop/coffee_output.csv")

# import weather data
weather123 = pd.read_csv("../data/weather/weather_data.csv")
```

```python
# TODO: Merge these two dataframes together

Coffee_weather = pd.merge (weather123,coffee,on=["subdivision","year"])
Coffee_weather.to_csv("../data/weather/Coffee_weather.csv", index=False)
```
